```python
import Wikiapi
import datetime

import requests
import json
import urllib
class Wikiapi:
    def getinfo(self,word):
        word=urllib.parse.quote(word)
        api="https://ja.wikipedia.org/w/api.php?action=query&format=json&titles={word}&prop=extracts&redirects=1&exchars=100&explaintext=1"
        api=api.format(word=word)
        r=requests.get(api)#APIにアクセス
        data=json.loads(r.text)
        pages=data["query"]["pages"]
        for page in pages:
            try:
                self.tweet=data["query"]["pages"][page]["extract"]#wikiの説明文を参照
                self.url="https://ja.wikipedia.org/wiki/"+word#urlもツイートに追加
                return self
            except KeyError:#記事なかった場合
                self.tweet="記事がありませんでした"
                return self
```

```python
def remove_link_tag(text):
    count = 0
    while text.find("[[") != -1:
        
        if count > 2000:
            print(text)
            break
            
        if text.find("[[") > text.find("]]"):
            if text.find("]]") != -1:
                text= text[:text.find("]]")] + text[text.find("]]")+2:]
            else:
                text= text[:text.find("[[")] + text[text.find("[[")+2:]
        else:
            tag = text[text.find("[["):text.find("]]")+2:]
            
            if tag.find("|") == -1:
                text = text.replace(tag,tag[tag.find("[[")+2:tag.find("]]"):])
            else:
                text = text.replace(tag,tag[tag.rfind("|")+1:tag.find("]]"):])
        count += 1
    return text

def remove_format_tag(text):
    count = 0
    while text.find("{{") != -1:
        
        if count > 2000:
            print(text)
            break
            
        if text.find("{{") > text.find("}}"):
            if text.find("}}") != -1:
                text= text[:text.find("}}")] + text[text.find("}}")+2:]
            else:
                text= text[:text.find("{{")] + text[text.find("{{")+2:]
        else:
            tag = text[text.find("{{"):text.find("}}")+2:]
            
            if tag.find("|") == -1:
                text = text.replace(tag,tag[tag.find("{{")+2:tag.find("}}"):])
            else:
                text = text.replace(tag,tag[tag.rfind("|")+1:tag.find("}}"):])
        count += 1
    return text
```

```python
category_per_count = 10000

wiki_id_list = []
category_id_list = []
company_list = []
text_list = []

category_dic = dict.fromkeys(list(category_pd["category"]),0)

file_name = './jawiki-latest-pages-articles.xml'

def get_title2(wiki_id):
    title = ""
    r = requests.get("https://ja.wikipedia.org/w/api.php?format=xml&action=query&pageids=" + str(wiki_id))
    soup = BeautifulSoup(r.content, "html.parser")
    try:
        title = soup.page['title']
        return title
    
    except:
        return ''
        
    
def get_title(s):
    s= re.sub(r'\s','',s)
    title = ''
    m = re.search('<title>([^<>]+)</title>',s)
    if m:
        #if m.group(1) not in train_company_list:
        title = m.group(1)
        
    #if re.search(r'\|社名\=(.*?)\|',s) != None:
    #    title_row = re.search(r'\|社名\=(.*?)\|',s).group(1)
    #elif re.search(r'\|社名\=(.*?)$',s) != None:
    #    title_row = re.search(r'\|社名\=(.*?)$',s).group(1)
    #else:return ''
    
    #if title_row.find(title[:3:].strip()) == -1 and title_row.find(title[-3::].strip()) == -1:
    #    return ''
    return title   


def get_wiki_id(s):
    wiki_id = ''
    m=re.search('<id>([^<>]+)</id>',s)
    if m and m.group(1) in company_id_list:
        wiki_id = m.group(1)
    return wiki_id

def get_category_id(s):
    
    #r = requests.get("https://ja.wikipedia.org/wiki/特別:データ書き出し/" + title)
    #soup = BeautifulSoup(r.content, "html.parser")
    #text = str(soup.select("text")[0])

    cleaned_text = remove_link_tag(s)
    cleaned_text = re.sub(r'\s','',cleaned_text)
    
    s=re.sub(r'\s','',s)
    category_id = ''
    jigyo = ''
    
    
    if re.search(r'\|事業内容\=(.*?)\n',cleaned_text) != None:
        jigyo = re.search(r'\|事業内容\=(.*?)\|',cleaned_text).group(1)
    elif re.search(r'\|事業内容\=(.*?)\|',cleaned_text) != None:
        jigyo = re.search(r'\|事業内容\=(.*?)\|',cleaned_text).group(1)
    elif re.search(r'\|事業内容\=(.*?)$',cleaned_text) != None:
        jigyo = re.search(r'\|事業内容\=(.*?)$',cleaned_text).group(1)  
    jigyo += "。"
    
    
    
    if re.search(r'\|業種\=(.*?)\|',s) != None:
        category_id_row = re.search(r'\|業種\=(.*?)\|',s).group(1)
    elif re.search(r'\|業種\=(.*?)$',s) != None:
        category_id_row = re.search(r'\|業種\=(.*?)$',s).group(1)
    else:
        return '',jigyo
    
    if re.search(r'<!--',category_id_row) != None:
        return '',jigyo
    
    category_id_list = re.findall(r'\d{4}',category_id_row)
    if len(category_id_list) != 0:
        for category_id_temp in category_id_list:
            if category_id_temp in list(category_pd["category"]):
                category_id = category_id_temp
    else:
        for category in list(category_pd["中分類"]):
            if category in category_id_row:
                category_id = category_pd["category"][category_pd["中分類"] == category].values[0]
    
    return category_id,jigyo

def get_gaiyou(s,title):
    gaiyou = []
    find_start = False
    find_end = False
    
    for i in s.split("\n"):
        if i.find('==') != -1:
            if i.find('概要') != -1 or i.find('概要') != -1:
                find_start = True
                continue
        elif i.find("'''") != -1:
            if i.find(title[:2:].strip()) != -1 or i.find(title[:title.find("("):].strip()[:-2:]) != -1:
                ###
                temp_num = re.search(r"は、|\）は|\)は",i)
                if temp_num is not None:
                    i = i[temp_num.start()+2::]
                    if i[:1:] == "、":
                        i = i[1::]
                    gaiyou.append(i)
                    find_start = True
                    continue
                    
        if find_end:
            break
            
        if find_start:
            if i.find('==') != -1:
                if i.find('概要') == -1 and i.find('概説') == -1:
                    find_end = True
                    continue
            gaiyou.append(i)
    return gaiyou


def get_text(gaiyou_list,wiki_id,title):
    r = requests.get("https://ja.wikipedia.org/?curid=" + str(wiki_id))
    soup = BeautifulSoup(r.content, "html.parser")
    for i in soup.select("p"):
        if title in i:
            gaiyou_list.append(i.get_text())
    
    return gaiyou_list,soup

    

def get_data(s):
    
    wiki_id = get_wiki_id(s)
    if wiki_id == '':
        #print("wiki_id")
        return
    
    title = get_title(s)
    if title == '':
        print("title:" + wiki_id)
        return
    
    category_id,jigyo = get_category_id(s)
    if category_id == '' or category_id is None:
        print("category:" + wiki_id)
        return
    #elif category_dic[category_id] >= category_per_count:
    #    print("category:" + wiki_id)
    #    return
    
    gaiyou_list = get_gaiyou(s,title)
    if len(gaiyou_list) == 0:
        gaiyou_list,soup = get_text(gaiyou_list,wiki_id,title)
        if len(gaiyou_list) == 0:
            print("gaiyou:" + wiki_id)
            return

    gaiyou_list.append(jigyo)
    gaiyou = ''.join(gaiyou_list)
    gaiyou = clean_text(gaiyou)
    
    if len(gaiyou) < 20:
        print("gaiyou:" + wiki_id)
        return
    
    
    category_dic[category_id] = category_dic[category_id] + 1
    
    wiki_id_list.append(wiki_id)
    category_id_list.append(category_id)
    company_list.append(title)
    text_list.append(gaiyou)
    
    f = open("./last_data/" + wiki_id + ".txt","w",encoding="UTF8")
    f.write(gaiyou)
    f.close
    
    #print(len(company_list))


import_test_pd = pd.DataFrame()
with open(file_name,'r',encoding="utf-8") as f:
    count = 0
    s = ""
    limit = 100000
    for l in f:
        if "<page>" in l: s = ""
        elif "</page>" in l:get_data(s)
        else:s += l
        count += 1
        if count > 10000000000 or len(company_list) >=limit:
            print("break")
            break
import_test_pd["company"] = company_list
import_test_pd["wiki_id"] = wiki_id_list
import_test_pd["category_id"] = category_id_list
import_test_pd["text"] = text_list
import_test_pd
```
